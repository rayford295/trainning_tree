```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = '/content/LosAngelesTrees_test.csv'

# 读取CSV文件
df = pd.read_csv(file_path)

# 计算不同GENUS的数量
unique_genera_count = df['GENUS'].nunique()

# 输出不同GENUS的数量
print(f'There are {unique_genera_count} unique genera.')

genera_counts = df['GENUS'].value_counts()

# 获取数量最多的前10个GENUS
top_10_genera = genera_counts.head(10)

# 获取数量最少的前10个GENUS
bottom_10_genera = genera_counts.tail(10)

print("Top 10 Genera:\n", top_10_genera)
print("\nBottom 10 Genera:\n", bottom_10_genera)

genera_counts = df['GENUS'].value_counts()

# 打印每个GENUS的计数
print(genera_counts)

# 计算每个GENUS出现的次数
genera_counts = df['GENUS'].value_counts()

# 绘制种类数的分布图
plt.figure(figsize=(10, 6))
genera_counts.plot(kind='bar')
plt.title('Distribution of Tree Genera')
plt.xlabel('Genus')
plt.ylabel('Count')
plt.show()

```
